```python
from detoxify import Detoxify


def get_toxicity(text: str) -> float:
    model = Detoxify(model_type="unbiased")
    result = model.predict(text)
    return result["toxicity"]
```

```python
get_toxicity("I hate you")
```

```python
get_toxicity("I love you")
```
